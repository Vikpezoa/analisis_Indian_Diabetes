```python
# Detectar valores nulos: Este Data Set no cuenta con valores ausentes o nulos
va=df.isnull().sum()
print('Valores ausentes: ', va)
```

```python
# Para detectar valores atipicos hicimos uso del grafico de caja y tambi√©n se hizo uso del rango intercuart√≠lo
import seaborn as sns

# Lista de nombres de columnas en tu DataFrame
column_names = df.columns

# Configurar el tama√±o de la figura
plt.figure(figsize=(12, 8))

# Iterar a trav√©s de las columnas y crear un gr√°fico de caja para cada campo
for i, column_name in enumerate(column_names):
  if column_name != column_names[8]:  # √öltimo campo variable categorica no recomendable boxplot
    plt.subplot(2, 4, i + 1)  # 2 filas, 5 columnas para los 10 gr√°ficos
    sns.boxplot(data=df[column_name])
    plt.title(column_name)
    plt.ylabel('Valor')
    plt.xlabel('')

plt.tight_layout()
plt.show()
```

```python
# Lista de nombres de columnas en tu DataFrame
column_names = df.columns

print("Old Shape: ", df.shape)

# Iterar a trav√©s de las columnas y realizar detecci√≥n y eliminaci√≥n de valores at√≠picos
for i, column_name in enumerate(column_names):
    if column_name != column_names[8]:  # Excluir columnas categ√≥ricas
        Q1 = np.percentile(df[column_name], 25, method='midpoint')
        Q3 = np.percentile(df[column_name], 75, method='midpoint')
        IQR = Q3 - Q1

        # Filtrar valores at√≠picos usando la condici√≥n booleana
        print(column_name)
        upper = df[column_name] >= (Q3 + 1.5 * IQR)
        print(np.where(upper))
        lower = df[column_name] <= (Q1 - 1.5 * IQR)
        print(np.where(lower))

        # Filtrar y obtener √≠ndices de filas para eliminar
        outlier_indices = df[upper | lower].index

        # Eliminar filas basadas en los √≠ndices de valores at√≠picos
        df.drop(outlier_indices, inplace=True)

print("New Shape: ", df.shape)
```

# 8. Realizar an√°lisis de correlaci√≥n y explicar que variables estan correlacionadas üîΩ

```python
import seaborn as sns

# Matriz de correlacion entre variables con un mapa de calor
corr_matrix = df.corr()

# Visualizar la matriz de correlacion con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlaci√≥n')
plt.show()
```


Observaciones: la variable edad y del n√∫mero de embarazos tiene un correlacion moderada positiva de 0.58 esto dice que a mayor edad tienden a presentar un numero mayor de embarazo.
La variable de insulina s√©rica de 2 horas con respecto a la variable del glosor del pliegue cut√°neo del triceps presenta una correlacion moderada de 0.48 esto quiere decir que a mayor glosor del pliegue cut√°neo del triceps mayor es el n√∫mero de la insulina S√©rica.
La variable de √≠ndice de Masa Corporal vs la funci√≥n de Pedrigr√≠ de Diabetes su correlaci√≥n es solo el 0.19, lo cual indica que √≠ndice de masa corporal no tiene dependencia con la funci√≥n de Pedrigr√≠ de Diabetes.

```python
plas = df['Concentraci√≥n de Glucosa en Plasma']
mass = df['√çndice de Masa Corporal']
class_var = df['Variable de Clase (0 o 1)']

#plt.scatter(plas, mass)
sns.scatterplot(x=plas, y=mass, hue=class_var)
plt.xlabel('Concentracion de glucosa en plasma')
plt.ylabel('Indice de masa corporal')
plt.title('Grafico de Dispersi√≥n: Glucosa en plasma vs √≠ndice de masa corporal')
plt.show()
```
