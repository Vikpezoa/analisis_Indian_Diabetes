# 4. Detectar y arreglar los siguientes conceptos 🔽

*   Valores Ausentes
*   Valores Atípicos


```python
# Detectar valores nulos: Este Data Set no cuenta con valores ausentes o nulos
va=df.isnull().sum()
print('Valores ausentes: ', va)
```

```python
# Para detectar valores atipicos hicimos uso del grafico de caja y también se hizo uso del rango intercuartílo
import seaborn as sns

# Lista de nombres de columnas en tu DataFrame
column_names = df.columns

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Iterar a través de las columnas y crear un gráfico de caja para cada campo
for i, column_name in enumerate(column_names):
  if column_name != column_names[8]:  # Último campo variable categorica no recomendable boxplot
    plt.subplot(2, 4, i + 1)  # 2 filas, 5 columnas para los 10 gráficos
    sns.boxplot(data=df[column_name])
    plt.title(column_name)
    plt.ylabel('Valor')
    plt.xlabel('')

plt.tight_layout()
plt.show()
```

```python
# Lista de nombres de columnas en tu DataFrame
column_names = df.columns

print("Old Shape: ", df.shape)

# Iterar a través de las columnas y realizar detección y eliminación de valores atípicos
for i, column_name in enumerate(column_names):
    if column_name != column_names[8]:  # Excluir columnas categóricas
        Q1 = np.percentile(df[column_name], 25, method='midpoint')
        Q3 = np.percentile(df[column_name], 75, method='midpoint')
        IQR = Q3 - Q1

        # Filtrar valores atípicos usando la condición booleana
        print(column_name)
        upper = df[column_name] >= (Q3 + 1.5 * IQR)
        print(np.where(upper))
        lower = df[column_name] <= (Q1 - 1.5 * IQR)
        print(np.where(lower))

        # Filtrar y obtener índices de filas para eliminar
        outlier_indices = df[upper | lower].index

        # Eliminar filas basadas en los índices de valores atípicos
        df.drop(outlier_indices, inplace=True)

print("New Shape: ", df.shape)
```

# 8. Realizar análisis de correlación y explicar que variables estan correlacionadas 🔽

```python
import seaborn as sns

# Matriz de correlacion entre variables con un mapa de calor
corr_matrix = df.corr()

# Visualizar la matriz de correlacion con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.show()
```


Observaciones: la variable edad y del número de embarazos tiene un correlacion moderada positiva de 0.58 esto dice que a mayor edad tienden a presentar un numero mayor de embarazo.
La variable de insulina sérica de 2 horas con respecto a la variable del glosor del pliegue cutáneo del triceps presenta una correlacion moderada de 0.48 esto quiere decir que a mayor glosor del pliegue cutáneo del triceps mayor es el número de la insulina Sérica.
La variable de índice de Masa Corporal vs la función de Pedrigrí de Diabetes su correlación es solo el 0.19, lo cual indica que índice de masa corporal no tiene dependencia con la función de Pedrigrí de Diabetes.

```python
plas = df['Concentración de Glucosa en Plasma']
mass = df['Índice de Masa Corporal']
class_var = df['Variable de Clase (0 o 1)']

#plt.scatter(plas, mass)
sns.scatterplot(x=plas, y=mass, hue=class_var)
plt.xlabel('Concentracion de glucosa en plasma')
plt.ylabel('Indice de masa corporal')
plt.title('Grafico de Dispersión: Glucosa en plasma vs índice de masa corporal')
plt.show()
```

Grafico de Dispersión: Glucosa en plasma vs índice de masa corporal
Se puede observar que se concentra la mayoria en los valores positivos en diabetes en aquellos con un indice de masa corporal mayor de 25 y menor a 50 con una concentración de glucosa mayor a 80

```python
plas = df['Concentración de Glucosa en Plasma']
nroEmb = df['Número de Embarazos']
class_var = df['Variable de Clase (0 o 1)']

sns.scatterplot(x=plas, y=nroEmb, hue=class_var)
plt.xlabel('Concentracion de Glucosa ')
plt.ylabel('Cantidad de veces de embarazo')
plt.title('Gráfico de Dispersion: Glucosa en plasma vs Cantidad de veces de embarazo')
plt.show()
```

En el gráfico se puede notar que no es clara en cuanto si hay un número exacto de veces de embarazo con relación a la variable glucosa en plasma.

```python
plas = df['Concentración de Glucosa en Plasma']
vble_edad = df['Edad']
class_var = df['Variable de Clase (0 o 1)']

sns.scatterplot(x=vble_edad, y=plas, hue=class_var)
plt.xlabel('Edad')
plt.ylabel('Concentracion de glucosa')
plt.title('Gráfico de Dispersion: Edad vs Glucosa en plasma')
plt.show()
```
